# **Relevant/Retained Features Description**
**Remittance_inflow:** This is the total amount of money received by a country from migrant working abroad and sending money back to their home country.            

**GDP_current_prices_US_Dollar:** Gross Domestic Product (GDP) at purchaser's prices is the sum of gross value added by all resident producers in the economy plus any product taxes and minus any subsidies not included in the value of the products. It is calculated without making deductions for depreciation of fabricated assets or for depletion and degradation of natural resources. Data are in current U.S. dollars. Dollar figures for GDP are converted from domestic currencies using single year official exchange rates. For a few countries where the official exchange rate does not reflect the rate effectively applied to actual foreign exchange transactions, an alternative conversion factor is used.                    

**Remittance_Percent_of_GDP**: Remittance percentage contribution to GDP.

**GDP_current_prices_PPP:** This indicator provides values for gross domestic product (GDP) expressed in current international dollars, converted by purchasing power parity (PPP) conversion factor. GDP is the sum of gross value added by all resident producers in the country plus any product taxes and minus any subsidies not included in the value of the products. PPP conversion factor is a spatial price deflator and currency converter that eliminates the effects of the differences in price levels between countries.        

**Inflation_average_consumer_prices_Index:** A consumer price index that measures the changes in the price of goods and services that households consumes which affects the real purchasing power of consumers’ income and their welfare. As the price of all goods and services do not change at the same time, a price index can only reflect an average movement and expressed as averages for year and not end-of-period data. Typically the index is assigned a unity value of 100. Inflation as measured by the consumer price index reflects the annual percentage change in the cost to the average consumer of acquiring a basket of goods and services that may be fixed or changed at specified intervals, such as yearly.                       

**Inflation_average_consumer_prices_Percent_change:** Inflation as measured by the consumer price index reflects the annual percentage change in the cost to the average consumer of acquiring a basket of goods and services that may be fixed or changed at specified intervals, such as yearly.  

**Human_Development_Index(HDI):** A composite index measuring average achievement in three basic dimensions of human development—a long and healthy life, knowledge and a decent standard of living.           

**Households_NPISHs_Final_consumption_expenditure_Annual_percent_growth:** Annual percentage growth of household final consumption expenditure per capita, which is calculated using household final consumption expenditure in constant 2010 prices and World Bank population estimates. Household final consumption expenditure (private consumption) is the market value of all goods and services, including durable products (such as cars, washing machines, and home computers), purchased by households.          

**Multidimensional_poverty_headcount_ratio_WorldBank_Percent_of_population:** The multidimensional poverty headcount ratio (World Bank) is the percentage of a population living in poverty according to the World Bank's Multidimensional Poverty Measure. The Multidimensional Poverty Measure includes three dimensions; monetary poverty, education, and basic infrastructure services, to capture a more complete picture of poverty.

**Poverty_headcount_ratio_at_societal_poverty_line_percent_of_population:** The poverty headcount ratio at societal poverty line is the percentage of a population living in poverty according to the World Bank's societal poverty line. The Societal Poverty Line is expressed in purchasing power adjusted 2017 U.S. dollars and defined as max(USD 2.15, USD 1.15 + USD 0.5*Median). This means that when the national median is sufficiently low, the Societal Poverty line is equivalent to the extreme poverty line, $2.15. For countries with a sufficiently high national median, the Societal Poverty Line grows as countries’ median income grows.

**Exploratory Data Analysis**

```python
df_remit=df[["Country", "Year", "Remittance_inflow_US_Dollar_Million"]]
df_remit
```

```python
Rem=df_remit[df["Year"]=='2023']
Rem
```

```python


plt.rcParams['figure.figsize'] = (10, 6)

ax = sns.barplot(y = Rem.Country, x = Rem.Remittance_inflow_US_Dollar_Million, data = Rem, color='g')

for i in ax.containers:
    ax.bar_label(i)

plt.gca().set_title("Remittance For Top 10 Recipient Countries In 2023 (US Dollar Million)")
plt.xlabel('Remittance inflow')
plt.ylabel('Country')


plt.rc('font', size=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

plt.show();
```

```python
df_remitGDP=df[["Country", "Year", "Remittance_Percent_of_GDP"]]
df_remitGDP
```

```python
remGDP=df_remitGDP[df["Year"]=='2023']
GDPrem=remGDP.sort_values(by='Remittance_Percent_of_GDP', ascending=False)
GDPrem
```

```python

plt.rcParams['figure.figsize'] = (10, 6)

ax = sns.barplot(y = GDPrem.Country, x = GDPrem.Remittance_Percent_of_GDP, data = GDPrem, color='y')

for i in ax.containers:
    ax.bar_label(i)

plt.gca().set_title("Remittance % of GDP For Top 10 Recipient Countries In 2023 (US Dollar Million)")
plt.xlabel('Remittance inflow')
plt.ylabel('Country')


plt.rc('font', size=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

plt.show();
```

```python
Remit=pd.DataFrame(df.groupby('Country')['Remittance_inflow_US_Dollar_Million'].sum().sort_values(ascending = False)).reset_index()[:10]


plt.rcParams['figure.figsize'] = (10, 6)

ax = sns.barplot(y = Remit.Country, x = Remit.Remittance_inflow_US_Dollar_Million, data = Remit, color='c')

for i in ax.containers:
    ax.bar_label(i)

plt.gca().set_title("Total Remittance For Top 10 Recepient Countries In 2023 From 2000 (US Dollar Million)")
plt.xlabel('Remittance inflow')
plt.ylabel('Country')


plt.rc('font', size=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

plt.show();
```

```python
Remit_PercentGDP=pd.DataFrame(df.groupby('Country')['Remittance_Percent_of_GDP'].sum().sort_values(ascending = False)).reset_index()[:10]
Remit_PercentGDP
```

```python
Total_remGDP=pd.DataFrame(df.groupby('Country')['Remittance_Percent_of_GDP'].sum().sort_values(ascending = False)).reset_index()[:10]


plt.rcParams['figure.figsize'] = (10, 6)

ax = sns.barplot(y = Total_remGDP.Country, x = Total_remGDP.Remittance_Percent_of_GDP, data = Remit_PercentGDP, color='r')

for i in ax.containers:
    ax.bar_label(i)

plt.gca().set_title("Total Remittance GDP Percentage Since 2000 For Top 10 Recepient Countries In 2023")
plt.xlabel('Remittance GDP Percentage')
plt.ylabel('Country')


plt.rc('font', size=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

plt.show();
```
